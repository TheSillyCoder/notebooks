```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
XLIM = 15
YLIM = 5
NPOINTS = 500
x = np.linspace(0, XLIM, NPOINTS)
y = np.linspace(-YLIM, YLIM, NPOINTS)

gridx, gridy = np.meshgrid(x, y, sparse=False)
```

```python
points = []
for i in range(len(gridx)):
    for j in range(len(gridx[i])):
        points.append(np.array([gridx[i][j], gridy[i][j]]))
points = np.array(points)
```

```python
slit1 = np.array([0, 1])
slit2 = np.array([0, -1])

dist1 = points - slit1
dist2 = points - slit2
A = 10
k = 10
wave1 = A*np.sin(k * (dist1[:,0]**2 + dist1[:,1]**2)**0.5)
wave2 = A*np.sin(k * (dist2[:,0]**2 + dist2[:,1]**2)**0.5)
# wave2 = A*np.sin(kx*dist2[:,0] + ky*dist2[:,1])
wave = wave1 + wave2
intensity = wave**2
```

```python
# plt.figure(figsize = (16,12))
plt.xlim(0, XLIM)
plt.ylim(-YLIM, YLIM)
plt.scatter(points[:,0], points[:,1], c=intensity, cmap="gray")
plt.scatter(slit1[0], slit1[1], color="red")
plt.scatter(slit2[0], slit2[1], color="red")
```
