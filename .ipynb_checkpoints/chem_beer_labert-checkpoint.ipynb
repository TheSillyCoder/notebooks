```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
```

```python
data1_x = np.array([80, 60, 40, 20, 10, 0])
data1_y = np.array([1.142380, 0.6124, 0.4576, 0.2856, 0.1789, 0.0143])
data1_u = 0.6433

data2_x = np.array([0.9, 0.8, 0.75, 0.667, 0.6, 0.5, 0.4, 0.333, 0.250, 0.200, 0.1])
data2_y = np.array([0.2525, 0.3147, 0.4070, 0.4538, 0.5230, 0.5793, 0.5341, 0.4778, 0.4472, 0.3984, 0.1944])
data2_u = 0.1944
```

```python
plt.plot(data1_x, data1_y, 'o')
def f(x, m, c):
    return m*x + c
popt, pcov = curve_fit(f, data1_x[1:], data1_y[1:])
x = np.linspace(0, 80, 100)
plt.plot(x, f(x, *popt))
unknown_concentration1 = (data1_u - popt[1]) / popt[0]
plt.axvline(unknown_concentration1)
plt.axhline(data1_u)
print(unknown_concentration1)
```

```python
fig, axes = plt.subplots(1,2, figsize=(20, 10))
ax = axes[1]
ax.plot(data2_x, data2_y, 'ko-')
ax.set_title("Job Plot of Absorbance versus Mole fraction of Fe(III)")
ax.set_xlabel("Mole fraction of Fe(III)")
ax.set_ylabel("Absorbance ($\lambda = 528 nm$)")
ax.annotate("Indicates that the reaction \n stoichiometry is 1:1", xy = [0.5,data2_y[5]], xytext = [0.6, 0.56], arrowprops=dict(facecolor='black', shrink=0.05, width=1))
ax = axes[0]
ax.plot(data1_x, data1_y, 'ko', label="Collected Absorbance Data")
ax.plot(x, f(x, popt[0], popt[1]), "k-", label=f"Linear Fit: f(x) = {popt[0]}x + {popt[1]}")
ax.set_title("Plot of Absorbance versus Concentration of Salicylate")
ax.legend(loc="upper left")
ax.scatter(unknown_concentration1, data1_u, color="black", s=100, marker="+")
ax.annotate(f"Unknown Concentration: {unknown_concentration1:.2f} mM", xy = [unknown_concentration1,data1_u], xytext = [30, 0.8], arrowprops=dict(facecolor='black', shrink=0.05, width=1))
ax.set_xlabel("Concentration of Salicylate [mM]")
ax.set_ylabel("Absorbance ($\lambda = 528 nm$)")
fig.savefig("chem_exp5_plot.pdf", dpi=300)
```
