```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy import constants
```

```python
data_file = open('data_planck.txt', 'r')
rows = data_file.readlines()
data_init = [row.split('\t') for row in rows]
data = [[],[]]
for i in range(len(data_init)):
    data[0].append(eval(data_init[i][0].strip()))
    data[1].append(eval(data_init[i][1].strip()))
```

```python
x = np.array(data[0])
y = np.array(data[1])
```

```python
def f(x, m, c):
    return (m * x) + c
popt, pcov = curve_fit(f, x, y)
```

```python
x_sample = np.linspace(4.5, 6.8, 50)
y_sample = f(x_sample, popt[0], popt[1])
```

```python
plt.style.use('default')
plt.figure(figsize=(16, 12))
# plt.xticks(tics)
plt.plot(x, y, 'ko', label="Coil 1", markersize=8)
# plt.plot(x, Super, 'g^-', label="Added readings of Coil1 and Coil 2", markersize=8)
plt.plot(x_sample, y_sample, 'b-', markersize=8)
plt.grid()
plt.xlabel("Frequency of incident light [sec^-1 x 10^14]", fontsize=17)
plt.ylabel("Stoppping Poential [V]", fontsize=17)
# plt.ylim(top=15.5)
# plt.legend(loc='upper right', fontsize=15)
plt.title("Determination of Planck's constant using Photoelectric Effect", fontsize=25)
plt.savefig("photoelec-plot.pdf", bbox_inches='tight', dpi=300)
```
