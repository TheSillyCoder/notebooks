```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
steps = 5000
N_itr = 10000
END = []
START = 0
for _ in range(N_itr):
    x = START
    for s in range(steps):
        step = np.random.choice([1,-1])
        x += step
    END.append(x)
```

```python
END = np.array(END)
plt.xlim(-300, 300)
plt.hist(END, density=True, bins=20)
x = np.linspace(-300, 300, 10000)
sig = np.sqrt(steps)
mu = 0
f = 1/(np.sqrt(2*np.pi) * sig) * np.exp(-((x - mu)/(2*sig))**2)
plt.plot(x, f)
```
