```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data_file = open('iso_data.txt', 'r')
```

```python
rows = data_file.readlines()
```

```python
data_init = [row.split('\t') for row in rows]
```

```python
data_init[len(data_init) - 1][2] = '0'
```

```python
data = [[],[],[]]
for i in range(len(data_init)):
    data[0].append(eval(data_init[i][0].strip()))
    data[1].append(eval(data_init[i][1].strip()))
    data[2].append(eval(data_init[i][2].strip()))
```

```python
v = data[0]
pH = data[1]
dpH_dV = data[2]
```

```python
plt.style.use('default')
plt.figure(figsize=(8,8))
plt.grid()
plt.plot(v, pH, 'ko-', markersize=8)
plt.xlabel("Volume of NaOH [ml]", fontsize=15)
plt.ylabel("pH", fontsize=15)
plt.savefig("isoelectronic-plot1.png")
```

```python
plt.style.use('default')
plt.figure(figsize=(8,8))
plt.grid()
plt.plot(v, dpH_dV, 'k-', markersize=8)
plt.xlabel("Volume of NaOH [ml]", fontsize=15)
plt.ylabel("d(pH)/dV", fontsize=15)
plt.savefig("isoelectronic-plot2.png")
```
