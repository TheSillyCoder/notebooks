```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data_file = open('data_helmholtz.dat', 'r')
```

```python
rows = data_file.readlines()
```

```python
data_init = [row.split('\t') for row in rows]
```

```python
data = [[],[],[],[],[],[]]
for i in range(len(data_init)):
    data[0].append(eval(data_init[i][0].strip()))
    data[1].append(eval(data_init[i][1].strip()))
    data[2].append(eval(data_init[i][2].strip()))
    data[3].append(eval(data_init[i][3].strip()))
    data[4].append(eval(data_init[i][4].strip()))
    data[5].append(eval(data_init[i][5].strip()))
```

```python
x = np.array(data[0]) - 26.5 - 5.5 - 15
# x = data[1]
B1 = data[2]
B2 = data[3]
Super = data[4]
B_H = data[5]
```

```python
plt.style.use('default')
plt.figure(figsize=(16, 12))
plt.plot(x, B1, 'bo-', label="Coil 1", markersize=8)
plt.plot(x, B2, 'rv-', label="Coil 2", markersize=8)
plt.plot(x, Super, 'g^-', label="Added readings of Coil1 and Coil 2", markersize=8)
plt.plot(x, B_H, 'ks-', label="Both Coil 1 and Coil 2", markersize=8)
plt.grid()
plt.xlabel("Distance from the centre of configuration [cm]", fontsize=15)
plt.ylabel("Magnetic Field Intensity [Gauss]", fontsize=15)
plt.ylim(top=15.5)
plt.legend(loc='upper right', fontsize=15)
plt.title("Magnetic Field Readings in Helmholtz Configuration", fontsize=20)
plt.savefig("helmholtz_plot.pdf")
```

```python
data_file.close()
```

```python
data_file = open('data_gt.txt','r')
rows = data_file.readlines()
rows.sort()
data_init = [row.split('\t') for row in rows]
```

```python
gt_data = [[],[]]
for i in range(len(data_init)):
    gt_data[0].append(eval(data_init[i][0].strip()))
    gt_data[1].append(eval(data_init[i][1].strip()))
```

```python
x_gt = np.array(gt_data[0]) - 26.5 - 15 - 8
B_gt = gt_data[1]
```

```python
plt.style.use('default')
plt.figure(figsize=(16, 12))
plt.plot(x_gt, B_gt, 'ko-', markersize=8)
plt.grid()
plt.xlabel("Distance from the centre of configuration [cm]", fontsize=15)
plt.ylabel("Magnetic Field Intensity [Gauss]", fontsize=15)
plt.title("Magnetic Field Readings when the coils are farther than Helmholtz Configuration", fontsize=20)
plt.savefig("greater_than_helmholtz.pdf")
```
